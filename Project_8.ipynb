```python
#1D Code

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

mean = [0, 0]
cov = [[1, 0.5], [0.5, 1.5]]

def pca(n):
  X_n = np.random.multivariate_normal(mean, cov, n)
  '''
  covar_1 = np.cov(np.transpose(X_n))
  w, v = np.linalg.eig(covar_1)
  '''
  pca = PCA(n_components = 2)
  pca.fit(X_n)
  w,v = pca.explained_variance_, pca.components_
  plt.scatter(X_n[:,0], X_n[:,1])
  plt.title('n='+str(n))
  plt.xlabel('X1')
  plt.ylabel('X2')
  plt.axis('equal')
  axes_1 = plt.axes()
  axes_1.quiver(0,0, w[0]*v[0,0], w[0]*v[0,1], units='xy', scale=1, color='purple')
  axes_1.quiver(0,0, w[1]*v[1,0], w[1]*v[1,1], units='xy', scale=1, color='purple')

plt.figure()
pca(10)
plt.figure()
pca(100)
plt.figure()
pca(1000)
plt.figure()
pca(10000)
```

```python
# 1E Code
from scipy.linalg import norm

mean = [0, 0]
cov = [[1, 0.5], [0.5, 1.5]]

def covari(n):
  X_n_1 = np.random.multivariate_normal(mean, cov, n)
  covar_1 = np.cov(np.transpose(X_n_1))
  norm_s = norm(covar_1)
  norm_1 = norm(cov)
  norm_diff = norm_s-norm_1
  return norm_diff

n = [10, 100, 1000, 10000, 100000, 1000000]

diff_l = []
for i in n:
  diff_l.append(covari(i))
plt.plot(n, diff_l)
plt.title('Difference Between Sample Covariance and True Covariance As Sample Size --> Infinity')
plt.xlabel('Sample Size')
plt.ylabel('Norm Difference Between Covariances')
```

```python
from sklearn import metrics
from scipy.spatial.distance import cdist

distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 10)
 
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(gmm[['X_1','X_2']])
    kmeanModel.fit(gmm[['X_1','X_2']])
 
    distortions.append(sum(np.min(cdist(gmm[['X_1','X_2']], kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / gmm[['X_1','X_2']].shape[0])
    inertias.append(kmeanModel.inertia_)
 
    mapping1[k] = sum(np.min(cdist(gmm[['X_1','X_2']], kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / gmm[['X_1','X_2']].shape[0]
    mapping2[k] = kmeanModel.inertia_
plt.subplot(2,2,1)
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion (GMM)')
plt.subplot(2,2,3)
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia (GMM)')

distortions_1 = []
inertias_1 = []
mapping1_1 = {}
mapping2_1 = {}
K_1 = range(1, 10)

for k in K_1:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(swissroll[['X_1', 'X_2']])
    kmeanModel.fit(swissroll[['X_1', 'X_2']])
 
    distortions_1.append(sum(np.min(cdist(swissroll[['X_1', 'X_2']], kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / swissroll[['X_1', 'X_2']].shape[0])
    inertias_1.append(kmeanModel.inertia_)
 
    mapping1[k] = sum(np.min(cdist(swissroll[['X_1', 'X_2']], kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / swissroll[['X_1', 'X_2']].shape[0]
    mapping2[k] = kmeanModel.inertia_
plt.subplot(2,2,2)
plt.plot(K_1, distortions_1, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion (Swissroll)')
plt.subplot(2,2,4)
plt.plot(K_1, inertias_1, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia (Swissroll)')
plt.tight_layout()
print('For GMM, at k of 4, the elbow method shows the distortion and inertia values decrease at a linear rate, which indicates an optimal k value.')
print('For Swissroll, the elbow method shows the distortion and inertia values never decrease at a linear rate, which indicates K-Means is not a good fit for this dataset.')
```
